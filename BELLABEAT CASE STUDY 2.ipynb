```python
# FIRST WE WILL RENAME THE DATE COLUMN 

#The date column is named different across the datasets, so we will rename it so it can be easier when we merge the datasets later 
```

```python
daily_activity = daily_activity.rename(columns={'ActivityDate': 'Date'})

sleep_day = sleep_day.rename(columns={'SleepDay': 'Date'})

hourly_steps = hourly_steps.rename(columns={'ActivityHour' : 'Time'})

hourly_calories = hourly_calories.rename(columns={'ActivityHour' : 'Time'})
```

```python
#CHANGING DATA FORMATTING 

# So looking at the outputs above we can see that the date column in all dataframes are of type objects. 
# We need to change the data type to determine to perform calculations 
```

```python
daily_activity['Date'] = pd.to_datetime(daily_activity['Date'])

sleep_day['Date'] = pd.to_datetime(sleep_day['Date'])

hourly_steps['Time'] = pd.to_datetime(hourly_steps['Time'])

hourly_calories['Time'] = pd.to_datetime(hourly_calories['Time'])
```

```python
#Aggregate is a function where the values of multiple rows are grouped together to form a single summary value

daily_activity.agg(
    {'TotalSteps': ['mean', 'min', 'max'],  
     'Calories': [ 'mean', 'min', 'max'],
    })
```

```python
# The average of total steps by the users is 7638 steps . according to a study conducted in 2011 by BMC/BIOMED CENTRAL , taking 
# 10000 steps a day is a reasonable target for healthy adults helping reduce certain health conditions such as high blood 
# pressure and heart disease. In order to compare daily steps to an activity level the following categories can be considered 

# Inactive - Less than 5000 steps/day
# Average - Between 7500 and 9999 steps/day
# Active - More than 12500 steps/day

# Given this information above we can conclude that our users fall into the average category with this in mind it becomes clear 
# that the users should improve their amount of dualy steps for optimal results regarding their health and well being.
```
